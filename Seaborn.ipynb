```python
# saturation value for lightness and darkness
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], ci= 50, n_boot= 2, palette='icefire', errcolor = "g", errwidth = 12)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data = var, hue="sex",
            hue_order=["Female", "Male"])
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data = var, hue="sex",
            hue_order=["Female", "Male"], orient = "h", palette='icefire')
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], capsize = 0.3)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], dodge=False)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], saturation = 100, alpha = 0.9)
plt.show()
```

```python
sns.set(style = "darkgrid")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], saturation = 100, alpha = 0.9)
plt.show()
```

```python
sns.set(style = "whitegrid")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data = var, hue="sex",
            order= order_1, hue_order=["Female", "Male"], saturation = 100, alpha = 0.9)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Heatmap Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=["attnr"], axis = 1).head(10)
x
```

```python
sns.heatmap(x)
plt.show()
```

```python
# change size of map using vmin and vmax 
# cmap change the color
# annot for value

```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin = 0, vmax=10, cmap='PuOr',annot=True)
plt.show()
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4",]])
ar
```

```python
sns.heatmap(var1, vmin = 0, vmax=10, cmap='PuOr', annot=ar,fmt="s", 
            linewidth=5, linecolor="b")
plt.show()
```

```python
# y ={"fontsize":10, "color":"r"} and annot_kws = y
sns.heatmap(var1, vmin = 0, vmax=10, cmap='PuOr', annot=ar,fmt="s", 
            linewidth=5, linecolor="b",cbar=False)
plt.show()
```

```python
# xticklabels for x axis
sns.heatmap(var1, vmin = 0, vmax=10, cmap='PuOr', annot=ar,fmt="s", 
            linewidth=5, linecolor="b",cbar=False, xticklabels=False)
plt.show()
```

```python
v = sns.heatmap(var1, vmin = 0, vmax=10, cmap='PuOr', annot=ar,fmt="s", 
            linewidth=5, linecolor="b",cbar=False, 
            xticklabels=False, yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
sns.set(font_scale=3)
plt.show()
```

# Facet Grid

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
